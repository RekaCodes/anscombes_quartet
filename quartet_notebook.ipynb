# Anscombe's Quartet
<i>Illustrating the importance of data visualization.</i> Made with `python` and `jupyter`.

___

### Overview

In 1973, English statistician Francis Anscombe created 'Anscombe's Quartet', a collection of four datasets each comprised of 11 (x,y) points. The same year, an article in the *American Statistician* journal featuring this dataset describes Anscombe's intent as a counter to the industry impression at the time that "numerical calculations are exact, but graphs are rough."

His construction of these datasets illustrates both the importance of data visualization as well as the effect of influential observations (such as outliers and duplicates) have on a dataset.

### Original Data

The original datasets include (x,y) values that share very similar statistical properties. In fact, the x-values for three of the four datasets are identical. 

See the dataset below:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv('Anscombe_quartet_data.csv')
```

```python
df1 = data[['x123', 'y1']]
df1.columns=['x', 'y']

df2 = data[['x123', 'y2']]
df2.columns=['x', 'y']

df3 = data[['x123', 'y3']]
df3.columns=['x', 'y']

df4 = data[['x4', 'y4']]
df4.columns=['x', 'y']

df = pd.concat([df1, df2, df3, df4], axis=1, keys=['I', 'II', 'III', 'IV'])
```

```python
df.style.set_caption('Raw Data')
```

### Summary Statistics
We get a better idea of the relationships between values by viewing the summary statistics.

```python
df.describe().style.set_caption('Summary Statistics')
```

### Data Visualization

Based on the data and summary statistics, one may mistake these as being similar datasets. However, when we see them visually, we see that they are completely different and tell very different stories.

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8))

fig.suptitle("Anscombe's Quartet", fontsize=20)

axes[0,0].set_title('Dataset I', fontsize=14)
axes[0,0].scatter(x=df1.x, y=df1.y)
axes[0,0].set_xlim([0, 20])
axes[0,0].set_ylim([0, 20])

axes[0,1].set_title('Dataset II', fontsize=14)
axes[0,1].scatter(x=df2.x, y=df2.y)
axes[0,1].set_xlim([0, 20])
axes[0,1].set_ylim([0, 20])

axes[1,0].set_title('Dataset III', fontsize=14)
axes[1,0].scatter(x=df3.x, y=df3.y)
axes[1,0].set_xlim([0, 20])
axes[1,0].set_ylim([0, 20])

axes[1,1].set_title('Dataset IV', fontsize=14)
axes[1,1].scatter(x=df4.x, y=df4.y)
axes[1,1].set_xlim([0, 20])
axes[1,1].set_ylim([0, 20])

fig.set_facecolor('linen')
fig.tight_layout()

plt.show()
```

### Summary

At first, the original data looks incredibly similar (remember the x-values are *identical* in three of the four). Even the summary statistics show nearly identical values.

When we visualize the data we see how different the datasets are from one another. It is only then that we see each tells its own very different story.
